```python
import pandas as pd

large = pd.read_csv("/content/submission_xlmrobert-large_5.csv")
large.rename(columns = {'label' : 'large'}, inplace = True)
base = pd.read_csv("/content/ensemble_xlmrobertbasemodels.csv")
base.rename(columns = {'label' : 'base'}, inplace = True)
xlnet = pd.read_csv("/content/ensemble_xlnetmodels.csv")
xlnet.rename(columns = {'label' : 'xlnet'}, inplace = True)
```

```python
one = pd.merge(large, base, on='id', how='inner')
df = pd.merge(xlnet, one, on='id', how='inner')
df
```

```python
df['label'] = df[['large', 'base', 'xlnet']].apply(lambda x: x.value_counts().idxmax(), axis=1)
```

```python
df[["id", "label"]].to_csv("/content/submission.csv", index=False)
```
